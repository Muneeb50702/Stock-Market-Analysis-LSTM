```python
import pandas as pd
df = pd.read_csv('to_be_done.csv')
df.columns
```

```python
df.head(10)
```

```python
df = df.dropna()
```

```python
cols = ['Ticker_AAPL', 'Ticker_AMZN', 'Ticker_BRK-B', 'Ticker_GOOGL', 
        'Ticker_JPM', 'Ticker_META', 'Ticker_MSFT', 'Ticker_NVDA', 
        'Ticker_TSLA', 'Ticker_V']

df[cols] = df[cols].astype(int)
df.head()
```

```python
del df['Year']
df.head()
```

```python
del df['Weekday_sin'], df['Weekday_cos']
```

```python
sum(df['Ticker_TSLA'] == 1)
```

```python
#Splitting specific companies data's into separate dataset to avoid Mixture while splitting
Ticker_AAPL = df[df['Ticker_AAPL'] == 1]
Ticker_AMZN = df[df['Ticker_AMZN'] == 1]
Ticker_BRKB = df[df['Ticker_BRK-B'] == 1]
Ticker_GOOGL = df[df['Ticker_GOOGL'] == 1]
Ticker_JPM = df[df['Ticker_JPM'] == 1]
Ticker_META = df[df['Ticker_META'] == 1]
Ticker_MSFT = df[df['Ticker_MSFT'] == 1]
Ticker_NVDA = df[df['Ticker_NVDA'] == 1]
Ticker_TSLA = df[df['Ticker_TSLA'] == 1]
Ticker_V = df[df['Ticker_V'] == 1]
```

```python
feature_cols = ['Open', 'Volume', 'SMA_10', 'SMA_50', 'EMA_20',
               'MACD', 'MACD_signal', 'RSI', 'BB_upper','BB_lower',
               'slowk', 'slowd', 'Ticker_AAPL', 'Ticker_AMZN',
               'Ticker_BRK-B', 'Ticker_GOOGL', 'Ticker_JPM', 'Ticker_META',
               'Ticker_MSFT', 'Ticker_NVDA', 'Ticker_TSLA', 'Ticker_V',
               'Month_sin', 'Month_cos', 'Day_sin', 'Day_cos']

target_cols = ['High', 'Low', 'Close']
time_steps = 10
```

```python
def create_sequences(data, feature_cols, target_cols, time_steps):
    X, Y = [], []
    for i in range(len(data) - time_steps):
        X.append(data[feature_cols].iloc[i:i + time_steps].values)
        Y.append(data[target_cols].iloc[i + time_steps].values)
    return X, Y
```

```python
company_dfs = [Ticker_AAPL, Ticker_AMZN, Ticker_BRKB, Ticker_GOOGL,
               Ticker_JPM, Ticker_META, Ticker_MSFT, Ticker_NVDA,
               Ticker_TSLA, Ticker_V]  

X_train_all, Y_train_all = [], []
X_test_all, Y_test_all = [], []

for df_company in company_dfs:
    df_company = df_company.copy()

    split = int(len(df_company) * 0.8)
    df_train = df_company.iloc[:split]
    df_test = df_company.iloc[split:]

    X_train, Y_train = create_sequences(df_train, feature_cols, target_cols, time_steps)
    X_test, Y_test = create_sequences(df_test, feature_cols, target_cols, time_steps)

    X_train_all.extend(X_train)
    Y_train_all.extend(Y_train)
    X_test_all.extend(X_test)
    Y_test_all.extend(Y_test)
```

```python
import numpy as np

X_train = np.array(X_train_all)
Y_train = np.array(Y_train_all)
X_test = np.array(X_test_all)
Y_test = np.array(Y_test_all)

print("X_train shape:", X_train.shape)  # (samples, time_steps, features)
print("Y_train shape:", Y_train.shape)  # (samples, targets)
print("X_test shape:", X_test.shape)
print("Y_test shape:", Y_test.shape)
```

```python
from sklearn.preprocessing import MinMaxScaler
import pickle

samples, timesteps, features = X_train.shape
samples1, timesteps1, features1 = X_test.shape
ysam, yfea = Y_train.shape
ysam1, yfea1 = Y_test.shape

# Reshape
X_train_reshaped = X_train.reshape(-1, features)
X_test_reshaped = X_test.reshape(-1, features)

y_train_reshaped = Y_train.reshape(-1, yfea)
y_test_reshaped = Y_test.reshape(-1, yfea1)

# Initialize scalers
scalary = MinMaxScaler(feature_range=(0, 1))
scaler = MinMaxScaler(feature_range=(0, 1))

# Fit only on training data
X_train_scaled = scaler.fit_transform(X_train_reshaped)
X_test_scaled = scaler.transform(X_test_reshaped)  # Changed here

Y_train_scaled = scalary.fit_transform(y_train_reshaped)
Y_test_scaled = scalary.transform(y_test_reshaped)  # Changed here

# Reshape back to original
X_train_scaled = X_train_scaled.reshape(samples, timesteps, features)
X_test_scaled = X_test_scaled.reshape(samples1, timesteps1, features1)
Y_train_scaled = Y_train_scaled.reshape(ysam, yfea)
Y_test_scaled = Y_test_scaled.reshape(ysam1, yfea1)

# Save the X scaler (optional to save Y scaler too)
with open('x_scaler_minmax_latest.pkl', 'wb') as f:
    pickle.dump(scaler, f)

with open('y_scaler_minmax_latest.pkl', 'wb') as f:
    pickle.dump(scalary, f)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.optimizers import Adam

model = Sequential()
model.add(LSTM(units=64, return_sequences=True, input_shape=(X_train.shape[1], X_train.shape[2])))
model.add(Dropout(0.2))
model.add(LSTM(units=32, return_sequences=False))
model.add(Dropout(0.2))
model.add(Dense(units=3))  # Output layer with 3 units (for High, Low, Close)

# Compile the model
model.compile(optimizer=Adam(learning_rate=0.001), loss='mean_squared_error')

# Train the model
model.fit(X_train_scaled, Y_train_scaled, epochs=20, batch_size=32, validation_data=(X_test_scaled, Y_test_scaled))
```

```python
test_loss = model.evaluate(X_test_scaled, Y_test_scaled)
print(f"Test Loss: {test_loss}")
```

```python
import matplotlib.pyplot as plt
predictions = model.predict(X_test_scaled)

with open('x_scaler_minmax_latest.pkl', 'rb') as f:
    x_scaler = pickle.load(f)

with open('y_scaler_minmax_latest.pkl', 'rb') as f:
    y_scaler = pickle.load(f)
    
predictions_original = y_scaler.inverse_transform(predictions)
Y_test_original = y_scaler.inverse_transform(Y_test_scaled)

plt.figure(figsize=(14, 5))
plt.plot(Y_test_original[:, 0], color='blue', label='Actual High')
plt.plot(predictions_original[:, 0], color='red', label='Predicted High')
plt.title('High Price - Actual vs Predicted')
plt.xlabel('Time Step')
plt.ylabel('Price')
plt.legend()
plt.show()

# Plot for 'Low'
plt.figure(figsize=(14, 5))
plt.plot(Y_test_original[:, 1], color='blue', label='Actual Low')
plt.plot(predictions_original[:, 1], color='red', label='Predicted Low')
plt.title('Low Price - Actual vs Predicted')
plt.xlabel('Time Step')
plt.ylabel('Price')
plt.legend()
plt.show()

# Plot for 'Close'
plt.figure(figsize=(14, 5))
plt.plot(Y_test_original[:, 2], color='blue', label='Actual Close')
plt.plot(predictions_original[:, 2], color='red', label='Predicted Close')
plt.title('Close Price - Actual vs Predicted')
plt.xlabel('Time Step')
plt.ylabel('Price')
plt.legend()
plt.show()
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

mae = mean_absolute_error(Y_test_scaled, predictions)
print(f"Mean Absolute Error (MAE): {mae}")

mse = mean_squared_error(Y_test_scaled, predictions)
print(f"Mean Squared Error (MSE): {mse}")

rmse = np.sqrt(mse)
print(f"Root Mean Squared Error (RMSE): {rmse}")

r2 = r2_score(Y_test_scaled, predictions)
print(f"R-squared (R²): {r2}")
```
